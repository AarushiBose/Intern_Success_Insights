```python
#6. How does the GPA vary among different colleges?
```

```python
# Sorting data by CGPA in descending order for top 5 results
colleges = df.head(5)

# Line plot
plt.figure(figsize=(12, 8))
plt.plot(colleges["College Name"], colleges["CGPA"], marker='o')

plt.xlabel("College Name")
plt.ylabel("CGPA")
plt.title("College vs GPA")
plt.xticks(rotation=70)
plt.tight_layout()
plt.show()
```

```python
# Calculate the average CGPA for each college
college_avg_cgpa = df.groupby('College Name')['CGPA'].mean()

# Sort the colleges by average CGPA in descending order and select the top 5
top_5_colleges = college_avg_cgpa.nlargest(5)

# Create a bar chart to visualize the average CGPA for the top 5 colleges
colors = ['blue', 'pink', 'lightgreen', 'orange', 'yellow']
plt.figure(figsize=(20,14))
top_5_colleges.plot(kind="bar", color=colors)
plt.xlabel("College Name")
plt.ylabel("Average CGPA")
plt.title("Average CGPA Among Different Colleges (Top 5 Highest CGPA Colleges)")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

```

```python
#7. Are there any outliers in the quantity(number of courses completed) attribute?
```

```python
attribute_name = 'Quantity'

# Calculating IQR (Interquartile Range) for the attribute
Q1 = df[attribute_name].quantile(0.25)
Q3 = df[attribute_name].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for identifying outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find the outliers
outliers = df[(df[attribute_name] < lower_bound) | (df[attribute_name] > upper_bound)]

# Check if there are any outliers
if len(outliers) > 0:
    print("There are outliers in the 'Quantity' attribute.")
    print("Outliers:")
    print(outliers)
else:
    print("There are no outliers in the 'Quantity' attribute.")

# Create a box plot to visualize the distribution of the 'Quantity' attribute
plt.figure(figsize=(8, 6))
plt.boxplot(df[attribute_name], vert=False)
plt.title("Box Plot of 'Quantity' Attribute")
plt.show()
```

```python
# 15. Which promotion channel brings in more number of students participations for the event?
```

```python
import pandas as pd
import matplotlib.pyplot as plt

promotion_channel_column = "How did you come to know about this event?"

# Count the number of participants for each promotion channel
channel_counts = df[promotion_channel_column].value_counts()

# bar chart to visualize the data
plt.figure(figsize=(20, 12))
channel_counts.plot(kind="barh", color='skyblue')
plt.xlabel("Promotion Channel")
plt.ylabel("Number of Participants")
plt.title("Event Participation by Promotion Channel")
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()

# Find the promotion channel with the most participants
most_participants_channel = channel_counts.idxmax()
print(f"The promotion channel with the most participants is: {most_participants_channel}")
```

```python
# 18. How many students know about the event from their college? Which of these top 5 colleges?
```

```python
# Filter the dataset to select students who came to know about the event from their college
college_condition = df['Specify in "Others" (how did you come to know about this event)'] == 'College'
college_students = df[college_condition]

# Count the number of students from each college
college_counts = college_students['College Name'].value_counts()

# Get the top 5 colleges with the highest number of students aware of the event
top_5_colleges = college_counts.head(5)

# Print the total number of students who know about the event from their college
total_students_from_college = len(college_students)
print(f"Total number of students who know about the event from their college: {total_students_from_college}")

# Print the top 5 colleges
print("\nTop 5 Colleges with the Most Students Aware of the Event:")
for idx, (college, count) in enumerate(top_5_colleges.iteritems(), start=1):
    print(f"{idx}. {college}: {count} students")
```
